```python
# 함수 (광역시도) 치면 인구수 가장 많은 행정구역 나오는 함수 만들기

def max_population(dosi):
    want = data.loc[data['광역시도'] == dosi, '인구수'].max()
    return data.loc[data['인구수'] == want, ['광역시도', '행정구역', '인구수']]
```

```python
max_population('경기도')
```

```python
def max_population(dosi):
    want = data.loc[data['광역시도']==dosi, '인구수'].max()
    return data.loc[data["인구수"]==want, ['행정구역', '인구수' ,'면적']]
```

```python
max_population('경기도')
```
