```python
#max() 함수
print(data.loc[data['광역시도'] == '강원도', 'population'].max())
print(data.loc[data['광역시도'] == '강원도', 'population'].min())
print(data.loc[data['광역시도'] == '강원도', 'population'].mean())
```

```python
max_pop = data.loc[data['광역시도'] == '강원도', \
                   'population'].max()
max_pop
```

```python
data.loc[data['population'] == max_pop,\
         ['광역시도','행정구역','population']]
```

```python
def get_maxpop_info(sido_name):
    max_pop = data.loc[data['광역시도'] == sido_name, \
                   'population'].max()
    return data.loc[data['population'] == max_pop,\
         ['광역시도','행정구역','population']]
```

```python
get_maxpop_info('부산광역시')
```

```python
get_maxpop_info('경상북도')
```

```python
# empty DataFrame 생성
max_pop_df = pd.DataFrame(columns=['광역시도','행정구역','population'])
max_pop_df
```

```python
for sido_name in data['광역시도'].unique():
    df = get_maxpop_info(sido_name)
    max_pop_df = max_pop_df.append(df)
```

```python
max_pop_df
```

```python
#reset index
max_pop_df = max_pop_df.reset_index()
max_pop_df
```

```python
# index column 삭제 : drop() 함수 
max_pop_df.drop('index', axis=1, inplace=True)
```

```python
max_pop_df
```

```python
#DataFrame을 excel file 로 저장
max_pop_df.to_excel('data/maxpop.xlsx',sheet_name='광역시도별 최대인구')
```

```python
#상관관계: 면적과 인구수의 상관관계
#corr() : -1 ~ 1 
#1에 가까울 수록 상관관계가 높다
#0에 가까울 수록 상관관계가 낮다.
data['면적'].corr(data['population'])
```

```python
seoul = data['광역시도'] == '서울특별시'
data.loc[seoul,'면적'].corr(data.loc[seoul,'population'])
```

```python
#서울특별시 인구수의 합계
data.loc[data['광역시도'] == '서울특별시', 'population'].sum()
```

```python
# groupby() 함수 사용
#광역시도별 인구수의 합계
#1. Series 객체의 groupby()
data['population'].groupby(data['광역시도']).sum()\
.sort_values(ascending=False)
```

```python
#2. DataFrame의 groupby() 함수 사용
data.groupby('광역시도')['population'].sum().sort_values(ascending=False)
```

```python
#광역시도별 면적의 합계(면적이 가장 넓은 순서로)
data.groupby('광역시도')['면적'].sum().sort_values(ascending=False)
```

```python
#광역시도별, 행정구역별 인구수의 합계
group_df = data.groupby(['광역시도','행정구역'])['population'].sum()
```
